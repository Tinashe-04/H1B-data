```python
# replace column names
data_2019.rename(columns={'Initial Approvals': 'Initial Approval', 'Initial Denials': 'Initial Denial', 'Continuing Approvals': 'Continuing Approval', 'Continuing Denials': 'Continuing Denial'}, inplace=True)
data_2019.info()
```

```python
def convert_to_int(string):
    '''
    Purpose: formatting the strings to remove commas and change data type to int data type
    Inputs: string to be formatted
    Returns: final_number, an int which is the result of the proper formatting
    '''
    final_number = 0

    # split the string up into components on either side of the comma
    first_list = string.split(',')

    # add all of the components together to produce one string
    number = ''.join(first_list)

    # convert that last string to an int
    final_number = int(number)

    return final_number

for i in data_2019.columns[2:6]:
    data_2019[i] = data_2019[i].apply(convert_to_int)

data_2019.info()
```

```python
from functools import reduce
dataframes = [data_2019, data_2020, data_2021, data_2022, data_2023]

data_2019_20 = pd.concat([data_2019, data_2020], ignore_index=True)
data_2019_20
```

```python
data_2020_21 = pd.concat([data_2019_20, data_2021], ignore_index=True)
data_2020_21
```

```python
data_2021_22 = pd.concat([data_2020_21, data_2022], ignore_index=True)
data_2021_22
```

```python
final_data = pd.concat([data_2021_22, data_2023], ignore_index=True)
final_data
```
